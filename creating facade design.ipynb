```python
#we had created 3 classes called SENSOR ! SMOKE ! LIGHTSS! 
#each class we have two METHODSS! 
#We are importing FACADE CLASS !to encapsulate all of them! 
class Sensor(object):
    def __init__(self):
        pass
    
    def sensorON(self):
        print('SENSOR ONNN!! ')
        
    def sensorOFF(self):
        print('SENSOR OFF')
        
        
class Smoke(object):
    def __init__(self):
        pass
    
    def smokeON(self):
        print('SMOKE ONNN!! ')
        
    def smokeOFF(self):
        print('SMOKE OFF')
    
    
class Lights(object):
    def __init__(self):
        pass
    
    def lightsON(self):
        print('LIGHTS ONNN!! ')
        
    def lightsOFF(self):
        print('LIGHTS OFF')
        

class Facade(object):
    def __init__(self):
        self._sensor = Sensor() #_ single underscore means private now we are encapsulation all the three classes! 
        self._smoke = Smoke()
        self._lights = Lights()
        
        
    def Emergency(self):
        self._sensor.sensorON()
        self._smoke.smokeON()
        self._lights.lightsON()
        
    def NoEmergency(self):
        self._sensor.sensorOFF()
        self._smoke.smokeOFF()
        self._lights.lightsOFF()
        
if __name__ == "__main__":
    facade = Facade()
    sensor = 100
    
    if sensor > 30:
        facade.Emergency()
        
    else:
        facade.NoEmergency()
        
    
        
    
    
```

```python
#Creating facade + singleton + Meta class! 

class Sensor(object):
    def __init__(self):
        pass
    
    def sensorON(self):
        print('SENSOR ONNN!! ')
        
    def sensorOFF(self):
        print('SENSOR OFF')
        
        
class Smoke(object):
    def __init__(self):
        pass
    
    def smokeON(self):
        print('SMOKE ONNN!! ')
        
    def smokeOFF(self):
        print('SMOKE OFF')
    
    
class Lights(object):
    def __init__(self):
        pass
    
    def lightsON(self):
        print('LIGHTS ONNN!! ')
        
    def lightsOFF(self):
        print('LIGHTS OFF')
        
class Meta1(type):
    """ SINGLTON DESIGN! """
    _instance = {}
    
    def __call__(cls,*args,**kwargs):
        
        """ creating the object! """
        if cls not in cls._instance:
            cls._instance[cls] = super(Meta1,cls).__call__(*args,**kwargs)
            return cls._instance[cls]
        
        
        

class Facade(metaclass = Meta1):
    def __init__(self):
        self._sensor = Sensor() #_ single underscore means private now we are encapsulation all the three classes! 
        self._smoke = Smoke()
        self._lights = Lights()
        
        
    def Emergency(self):
        self._sensor.sensorON()
        self._smoke.smokeON()
        self._lights.lightsON()
        
    def NoEmergency(self):
        self._sensor.sensorOFF()
        self._smoke.smokeOFF()
        self._lights.lightsOFF()
   

if __name__ == '__main__':
    facade = Facade()
    print(facade)
    facade1 = Facade()
    print(facade1)
  
```
